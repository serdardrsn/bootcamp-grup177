```python
import streamlit as st
import requests

st.title("🚑 Emergency Triage AI (KTAS Tahmini)")

# Chief Complaint
chief_complaint = st.text_input("🗣️ Başvuru Şikayeti (Chief Complaint)", placeholder="Örn: Göğüs ağrısı")

# Sayısal girişler
age = st.number_input("Yaş", min_value=0, max_value=120, step=1)
sbp = st.number_input("Sistolik Kan Basıncı (SBP)", min_value=0)
dbp = st.number_input("Diyastolik Kan Basıncı (DBP)", min_value=0)
hr = st.number_input("Nabız (HR)", min_value=0)
rr = st.number_input("Solunum Hızı (RR)", min_value=0)
bt = float(st.number_input("Vücut Isısı (BT)", min_value=30.0, format="%.1f"))

# Kategorik girişler
mental = st.selectbox("Bilinç Durumu", ["Uyanık", "Yorgun", "Tepkisiz"])
sex = st.selectbox("Cinsiyet", ["Erkek", "Kadın"])
arrival_mode = st.selectbox("Hastanın Geliş Şekli", ["Ambulans", "Yürüyerek", "Sedye ile"])
injury = st.selectbox("Travma Durumu", ["Evet", "Hayır"])
pain = st.slider("Ağrı Puanı (0-10)", 0, 10, step=1)

# Buton
if st.button("🧠 Tahmin Yap"):
    input_data = {
        "chief_complaint": chief_complaint,
        "age": int(age),
        "sbp": int(sbp),
        "dbp": int(dbp),
        "hr": int(hr),
        "rr": int(rr),
        "bt": float(bt),
        "mental": mental,
        "sex": sex,
        "arrival_mode": arrival_mode,
        "injury": injury,
        "pain": int(pain)
    }

    st.subheader("📤 Gönderilen Veri:")
    st.json(input_data)

    try:
        response = requests.post("http://127.0.0.1:8000/predict", json=input_data)
        result = response.json()

        st.success(f"✅ KTAS Tahmini: {result['ktas_tahmini']}")
        st.markdown("### 💬 Modelin Cevabı:")
        st.code(result['model_cevabi'], language="markdown")

    except Exception as e:
        st.error(f"❌ API bağlantısı başarısız: {e}")
```
