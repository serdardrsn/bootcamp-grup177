```python
import streamlit as st
import requests

st.title("üöë Emergency Triage AI (KTAS Tahmini)")

# Chief Complaint
chief_complaint = st.text_input("üó£Ô∏è Ba≈üvuru ≈ûikayeti (Chief Complaint)", placeholder="√ñrn: G√∂ƒü√ºs aƒürƒ±sƒ±")

# Sayƒ±sal giri≈üler
age = st.number_input("Ya≈ü", min_value=0, max_value=120, step=1)
sbp = st.number_input("Sistolik Kan Basƒ±ncƒ± (SBP)", min_value=0)
dbp = st.number_input("Diyastolik Kan Basƒ±ncƒ± (DBP)", min_value=0)
hr = st.number_input("Nabƒ±z (HR)", min_value=0)
rr = st.number_input("Solunum Hƒ±zƒ± (RR)", min_value=0)
bt = float(st.number_input("V√ºcut Isƒ±sƒ± (BT)", min_value=30.0, format="%.1f"))

# Kategorik giri≈üler
mental = st.selectbox("Bilin√ß Durumu", ["Uyanƒ±k", "Yorgun", "Tepkisiz"])
sex = st.selectbox("Cinsiyet", ["Erkek", "Kadƒ±n"])
arrival_mode = st.selectbox("Hastanƒ±n Geli≈ü ≈ûekli", ["Ambulans", "Y√ºr√ºyerek", "Sedye ile"])
injury = st.selectbox("Travma Durumu", ["Evet", "Hayƒ±r"])
pain = st.slider("Aƒürƒ± Puanƒ± (0-10)", 0, 10, step=1)

# Buton
if st.button("üß† Tahmin Yap"):
    input_data = {
        "chief_complaint": chief_complaint,
        "age": int(age),
        "sbp": int(sbp),
        "dbp": int(dbp),
        "hr": int(hr),
        "rr": int(rr),
        "bt": float(bt),
        "mental": mental,
        "sex": sex,
        "arrival_mode": arrival_mode,
        "injury": injury,
        "pain": int(pain)
    }

    st.subheader("üì§ G√∂nderilen Veri:")
    st.json(input_data)

    try:
        response = requests.post("http://127.0.0.1:8000/predict", json=input_data)
        result = response.json()

        st.success(f"‚úÖ KTAS Tahmini: {result['ktas_tahmini']}")
        st.markdown("### üí¨ Modelin Cevabƒ±:")
        st.code(result['model_cevabi'], language="markdown")

    except Exception as e:
        st.error(f"‚ùå API baƒülantƒ±sƒ± ba≈üarƒ±sƒ±z: {e}")
```
